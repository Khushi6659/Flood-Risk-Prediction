# Title: Flood risk prediction in India using machine learning

## Problem Statement and Description :
#### Floods in India cause severe damage due to unpredictable rainfall, rising river discharge,
#### and growing population pressure. Existing models mostly rely on basic factors like rainfall and water levels,
#### which limits accuracy. This project aims to build an enhanced flood risk prediction system using historical climate,
#### hydrological,geographical, and socio-environmental data. Along with core features (rainfall, river discharge,
#### water level,land cover, population density, soil type), it introduces new features such as Rainfall Anomaly,
#### Flood Risk Index,and Urbanization Pressure Score. By combining these, the system can provide early,
#### location-specific flood warnings, supporting disaster management authorities in reducing economic loss and
#### safeguarding communities.¶

## Data Transformation
#### Machine learning models cannot directly use categorical features like "Land Cover" and "Soil Type". Also, scaling is needed so no feature dominates others.
#### 1) Label Encoding: Converts text labels into numbers.
####    Example: Land Cover → Forest=0, Urban=1, Water Body=2, etc.

#### 2) Standard Scaling: Converts numeric features into a standard scale (mean=0, std=1).
####    Example: Rainfall ranges from 0–300, while Water Level is 0–10. Without scaling, rainfall would dominate.

```python
# Copy dataset
data = df.copy()

# Encode categorical variables
le_land = LabelEncoder()
le_soil = LabelEncoder()
data["Land Cover"] = le_land.fit_transform(data["Land Cover"])
data["Soil Type"] = le_soil.fit_transform(data["Soil Type"])

# Scale numerical features
scaler = StandardScaler()
num_cols = ["Rainfall (mm)", "Temperature (°C)", "Humidity (%)", 
            "River Discharge (m³/s)", "Water Level (m)", 
            "Elevation (m)", "Population Density"]

data[num_cols] = scaler.fit_transform(data[num_cols])

print("\nTransformed Data (first 5 rows):")
print(data.head())
```

## Feature Selection
#### We now check which features are most important for predicting Flood Occurrence.
#### SelectKBest with f_classif → Selects features most correlated with the target variable (Flood Occurred).

```python
# Split features and target
X = data.drop("Flood Occurred", axis=1)
y = data["Flood Occurred"]

# Select top 8 features
selector = SelectKBest(score_func=f_classif, k=8)
X_new = selector.fit_transform(X, y)

selected_features = X.columns[selector.get_support()]
print("\nSelected Features:", selected_features.tolist())
```

## Smote

```python
# smote_preprocessing.py
import pandas as pd
from sklearn.preprocessing import LabelEncoder, StandardScaler
from imblearn.over_sampling import SMOTE
import joblib

# ==============================
# Load Dataset
# ==============================
df = pd.read_csv("flood_risk_dataset_india.csv")
print(" Dataset Loaded:", df.shape)

# Encode categorical features
label_encoders = {}
for col in df.select_dtypes(include=['object']).columns:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])
    label_encoders[col] = le

# Feature Engineering
df["Rainfall_Anomaly"] = df["Rainfall (mm)"] - df["Rainfall (mm)"].rolling(7, min_periods=1).mean()
df["Flood_Risk_Index"] = (df["Rainfall (mm)"] * df["River Discharge (m³/s)"]) / (df["Elevation (m)"] + 1)
df["Humid_Temp_Interaction"] = df["Humidity (%)"] * df["Temperature (°C)"]
df["Population_Exposure"] = df["Population Density"] * df["Infrastructure"]
df["Elevation_River_Ratio"] = df["Elevation (m)"] / (df["River Discharge (m³/s)"] + 1)
df = df.fillna(0)

# Split X, y
X = df.drop(columns=["Flood Occurred"])
y = df["Flood Occurred"]

# Apply SMOTE
smote = SMOTE(random_state=42)
X_res, y_res = smote.fit_resample(X, y)

# Scale features
scaler = StandardScaler()
X_res_scaled = scaler.fit_transform(X_res)

print(" After SMOTE:", X_res_scaled.shape, y_res.shape)

# Save preprocessed dataset
joblib.dump((X_res_scaled, y_res, scaler, label_encoders), "smote_data.pkl")
print(" Saved balanced dataset as smote_data.pkl")
```

```python
# ============================
# Strong Stacking Pipeline for Flood Prediction
# Paste into Jupyter and run
# ============================

# If you need to install packages run in a cell first (uncomment if needed):
# !pip install xgboost catboost imbalanced-learn joblib

import warnings
warnings.filterwarnings("ignore")

import os
import time
import joblib
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split, StratifiedKFold, cross_val_score
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, roc_auc_score
from sklearn.feature_selection import SelectFromModel
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, StackingClassifier

# Try imports that might not be installed
use_xgb = True
use_cat = True
use_smote = True
try:
    from xgboost import XGBClassifier
except Exception as e:
    print("XGBoost not found — will skip XGBoost. Install with `pip install xgboost` to enable it.")
    use_xgb = False

try:
    from catboost import CatBoostClassifier
except Exception as e:
    print("CatBoost not found — will skip CatBoost. Install with `pip install catboost` to enable it.")
    use_cat = False

try:
    from imblearn.over_sampling import SMOTE
except Exception as e:
    print("imblearn (SMOTE) not found — will continue without SMOTE. Install with `pip install imbalanced-learn` to enable it.")
    use_smote = False

# -----------------------------
# 1) Load dataset
# -----------------------------
DATA_PATH = "flood_risk_dataset_india.csv"
assert os.path.exists(DATA_PATH), f"File not found at {DATA_PATH}"

df = pd.read_csv(DATA_PATH)
print("Loaded dataset:", df.shape)
display(df.head())

# -----------------------------
# 2) Quick checks
# -----------------------------
print("\nMissing values per column:")
print(df.isnull().sum())

print("\nTarget distribution (Flood Occurred):")
print(df["Flood Occurred"].value_counts(normalize=False))
print(df["Flood Occurred"].value_counts(normalize=True))

# -----------------------------
# 3) Feature engineering (adds more informative signals)
# -----------------------------
# Create engineered features from existing columns
df = df.copy()

# 3.1 rainfall anomaly (difference from overall mean)
df["Rainfall_Anomaly"] = df["Rainfall (mm)"] - df["Rainfall (mm)"].mean()

# 3.2 normalized risk index: rainfall * discharge / (elevation + 1)
df["Flood_Risk_Index_v2"] = (df["Rainfall (mm)"] * (df["River Discharge (m³/s)"] + 1)) / (df["Elevation (m)"] + 1)

# 3.3 humidity * temperature interaction
df["Humid_Temp"] = df["Humidity (%)"] * df["Temperature (°C)"]

# 3.4 population exposure (pop density * historical floods)
df["Population_Exposure"] = df["Population Density"] * (df["Historical Floods"] + 1)

# 3.5 elevation to water_level ratio
df["Elevation_WaterRatio"] = df["Elevation (m)"] / (df["Water Level (m)"] + 1)

# 3.6 discharge to rainfall ratio
df["Discharge_Rainfall_Ratio"] = (df["River Discharge (m³/s)"] + 1) / (df["Rainfall (mm)"] + 1)

# Replace any inf / na resulting from division
df.replace([np.inf, -np.inf], np.nan, inplace=True)
df.fillna(0, inplace=True)

print("\nAdded engineered features. New shape:", df.shape)

# -----------------------------
# 4) Encode categorical variables
# -----------------------------
categorical_cols = df.select_dtypes(include=["object"]).columns.tolist()
encoders = {}
for col in categorical_cols:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col].astype(str))
    encoders[col] = le
print("\nEncoded categorical columns:", categorical_cols)

# -----------------------------
# 5) Prepare X, y
# -----------------------------
target_col = "Flood Occurred"
assert target_col in df.columns, "Target column 'Flood Occurred' not found"

X = df.drop(columns=[target_col])
y = df[target_col].astype(int)

print("\nFeature count:", X.shape[1])

# -----------------------------
# 6) Feature selection using RandomForest importance (select top features)
# -----------------------------
# Quick RF to get importances - use small forest for speed
rf_for_fs = RandomForestClassifier(n_estimators=200, random_state=42, n_jobs=-1)
rf_for_fs.fit(X, y)
importances = pd.Series(rf_for_fs.feature_importances_, index=X.columns).sort_values(ascending=False)
print("\nTop feature importances:")
display(importances.head(15))

# Select features above median importance or top 12 (whichever bigger)
threshold = np.median(importances.values)
sel = SelectFromModel(rf_for_fs, threshold=threshold, prefit=True)
X_selected = sel.transform(X)
selected_features = X.columns[sel.get_support()].tolist()
print(f"\nSelected {len(selected_features)} features: {selected_features}")

# -----------------------------
# 7) SMOTE balancing (optional but recommended)
# -----------------------------
if use_smote:
    smote = SMOTE(random_state=42, n_jobs=-1)
    X_bal, y_bal = smote.fit_resample(X_selected, y)
    print("\nAfter SMOTE balanced shape:", X_bal.shape, y_bal.shape)
else:
    X_bal, y_bal = X_selected, y
    print("\nSMOTE not used — using original data.")

# -----------------------------
# 8) Scale features
# -----------------------------
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_bal)

# -----------------------------
# 9) Train-test split (stratified)
# -----------------------------
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y_bal, test_size=0.20, random_state=42, stratify=y_bal)
print("\nTraining samples:", X_train.shape[0], "Testing samples:", X_test.shape[0])

# -----------------------------
# 10) Define strong base learners
# -----------------------------
# tuned-ish parameters (kept reasonable for your laptop)
estimators = []

# Random Forest
rf = RandomForestClassifier(n_estimators=400, max_depth=12, random_state=42, n_jobs=-1)
estimators.append(("rf", rf))

# Gradient Boosting as fallback
gb = GradientBoostingClassifier(n_estimators=300, learning_rate=0.05, max_depth=5, random_state=42)
estimators.append(("gb", gb))

# XGBoost (if available)
if use_xgb:
    xgb_clf = XGBClassifier(n_estimators=300, learning_rate=0.05, max_depth=6, use_label_encoder=False, eval_metric="logloss", random_state=42, n_jobs=1)
    estimators.append(("xgb", xgb_clf))

# CatBoost (if available) — it handles categorical well but we've encoded already
if use_cat:
    cat_clf = CatBoostClassifier(iterations=300, learning_rate=0.05, depth=6, verbose=0, random_state=42)
    estimators.append(("cat", cat_clf))

print("\nBase estimators in stack:", [name for name, _ in estimators])

# -----------------------------
# 11) Build StackingClassifier with logistic meta-learner
# -----------------------------
final_est = LogisticRegression(max_iter=1000)
stack = StackingClassifier(estimators=estimators, final_estimator=final_est, n_jobs=-1, passthrough=False)

# -----------------------------
# 12) Cross-validated estimate (quick)
# -----------------------------
print("\nRunning 5-fold cross-validation on stacking (may take a few minutes)...")
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
start_time = time.time()
cv_scores = cross_val_score(stack, X_train, y_train, cv=cv, scoring="accuracy", n_jobs=-1)
print("CV accuracy scores:", np.round(cv_scores, 4))
print("CV mean accuracy:", np.round(cv_scores.mean(), 4))
print("Time elapsed (CV):", round(time.time() - start_time, 1), "s")

# -----------------------------
# 13) Fit stack on full training set & evaluate on test
# -----------------------------
print("\nFitting final stack on training set (this will take a minute)...")
start_time = time.time()
stack.fit(X_train, y_train)
fit_time = round(time.time() - start_time, 1)
print("Training finished in", fit_time, "s")

# Predict & evaluate
y_pred = stack.predict(X_test)
acc = accuracy_score(y_test, y_pred)
print("\n---- Test set results ----")
print("Accuracy:", round(acc, 4))
print(classification_report(y_test, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))

# ROC-AUC if possible (binary)
try:
    y_proba = stack.predict_proba(X_test)[:, 1]
    roc = roc_auc_score(y_test, y_proba)
    print("ROC-AUC:", round(roc, 4))
except Exception:
    pass

# -----------------------------
# 14) Feature importance (aggregate from RF if available)
# -----------------------------
try:
    rf_model = None
    for name, mod in estimators:
        if name == "rf":
            rf_model = mod
            break
    if rf_model is None:
        rf_model = RandomForestClassifier(n_estimators=300, random_state=42)
        rf_model.fit(X_train, y_train)
    else:
        rf_model.fit(X_train, y_train)
    fi = rf_model.feature_importances_
    feat_names = selected_features
    fi_series = pd.Series(fi, index=feat_names).sort_values(ascending=False)
    print("\nTop feature importances (from RandomForest):")
    display(fi_series.head(12))
    # Plot
    plt.figure(figsize=(8,6))
    sns.barplot(x=fi_series.values[:12], y=fi_series.index[:12])
    plt.title("Top feature importances")
    plt.show()
except Exception as e:
    print("Could not compute feature importances:", e)

# -----------------------------
# 15) Save artifacts
# -----------------------------
joblib.dump({"stack_model": stack, "scaler": scaler, "encoders": encoders, "selected_features": selected_features}, "flood_stack_artifacts.pkl")
print("\nSaved stack model + scaler + encoders to 'flood_stack_artifacts.pkl'")

# -----------------------------
# 16) Estimated runtime note for your laptop
# -----------------------------
print("\nEstimated run time on Intel i5 (8GB RAM gaming laptop):")
print("- CV (5-fold) + fit + prediction: ~2-8 minutes depending on XGBoost/CatBoost presence.")
print("- If XGBoost/CatBoost are missing, it's faster (~2-4 minutes).")
print("\nIf you want faster runs during development, reduce n_estimators or exclude heavy boosters.")
```

```python
# flood_risk_dl.py
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.decomposition import PCA
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping

# =========================
# Step 1: Load Dataset
# =========================
df = pd.read_csv("flood_risk_dataset_india.csv")
print("Dataset Loaded:", df.shape)

# Encode categorical columns
label_encoders = {}
for col in df.select_dtypes(include=['object']).columns:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])
    label_encoders[col] = le

# Separate features & target
X = df.drop("Flood Occurred", axis=1)
y = df["Flood Occurred"]

# =========================
# Step 2: Scale Features
# =========================
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# =========================
# Step 3: Dimensionality Reduction (PCA)
# =========================
pca = PCA(n_components=15)  # keep top 15 components
X_pca = pca.fit_transform(X_scaled)
print("PCA reduced shape:", X_pca.shape)

# =========================
# Step 4: Train-Test Split
# =========================
X_train, X_test, y_train, y_test = train_test_split(
    X_pca, y, test_size=0.2, random_state=42, stratify=y
)

# =========================
# Step 5: Deep Learning Model
# =========================
model = Sequential()
model.add(Dense(128, activation="relu", input_dim=X_train.shape[1]))
model.add(Dropout(0.3))
model.add(Dense(64, activation="relu"))
model.add(Dropout(0.3))
model.add(Dense(32, activation="relu"))
model.add(Dense(1, activation="sigmoid"))

model.compile(optimizer="adam", loss="binary_crossentropy", metrics=["accuracy"])

# Early stopping
es = EarlyStopping(monitor="val_loss", patience=10, restore_best_weights=True)

# Train
history = model.fit(
    X_train, y_train,
    validation_data=(X_test, y_test),
    epochs=100,
    batch_size=64,
    callbacks=[es],
    verbose=1
)

# =========================
# Step 6: Evaluation
# =========================
y_pred = (model.predict(X_test) > 0.5).astype("int32")
acc = accuracy_score(y_test, y_pred)
print("\nFinal Test Accuracy:", acc)
print("\nClassification Report:\n", classification_report(y_test, y_pred))
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred))
```
