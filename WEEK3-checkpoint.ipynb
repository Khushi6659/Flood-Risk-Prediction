## ML Model

```python
# ============================
# Strong Stacking Pipeline for Flood Prediction
# With Confusion Matrix Plot & Save
# ============================

# Install if needed (uncomment and run once in Jupyter):
# !pip install xgboost catboost imbalanced-learn joblib seaborn

import warnings
warnings.filterwarnings("ignore")

import os
import time
import joblib
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split, StratifiedKFold, cross_val_score
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, roc_auc_score
from sklearn.feature_selection import SelectFromModel
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, StackingClassifier

# Optional imports
use_xgb = True
use_cat = True
use_smote = True
try:
    from xgboost import XGBClassifier
except Exception:
    print("XGBoost not found, skipping...")
    use_xgb = False

try:
    from catboost import CatBoostClassifier
except Exception:
    print("CatBoost not found, skipping...")
    use_cat = False

try:
    from imblearn.over_sampling import SMOTE
except Exception:
    print("SMOTE not found, skipping...")
    use_smote = False

# -----------------------------
# 1) Load dataset
# -----------------------------
DATA_PATH = "flood_risk_dataset_india.csv"
assert os.path.exists(DATA_PATH), f"File not found at {DATA_PATH}"

df = pd.read_csv(DATA_PATH)
print("Loaded dataset:", df.shape)
print(df.head())

# -----------------------------
# 2) Feature engineering
# -----------------------------
df = df.copy()
df["Rainfall_Anomaly"] = df["Rainfall (mm)"] - df["Rainfall (mm)"].mean()
df["Flood_Risk_Index_v2"] = (df["Rainfall (mm)"] * (df["River Discharge (m³/s)"] + 1)) / (df["Elevation (m)"] + 1)
df["Humid_Temp"] = df["Humidity (%)"] * df["Temperature (°C)"]
df["Population_Exposure"] = df["Population Density"] * (df["Historical Floods"] + 1)
df["Elevation_WaterRatio"] = df["Elevation (m)"] / (df["Water Level (m)"] + 1)
df["Discharge_Rainfall_Ratio"] = (df["River Discharge (m³/s)"] + 1) / (df["Rainfall (mm)"] + 1)

df.replace([np.inf, -np.inf], np.nan, inplace=True)
df.fillna(0, inplace=True)
print("New shape after feature engineering:", df.shape)

# -----------------------------
# 3) Encode categorical variables
# -----------------------------
categorical_cols = df.select_dtypes(include=["object"]).columns.tolist()
encoders = {}
for col in categorical_cols:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col].astype(str))
    encoders[col] = le
print("Encoded categorical columns:", categorical_cols)

# -----------------------------
# 4) Prepare X, y
# -----------------------------
target_col = "Flood Occurred"
assert target_col in df.columns, "Target column 'Flood Occurred' not found"

X = df.drop(columns=[target_col])
y = df[target_col].astype(int)

# -----------------------------
# 5) Feature selection
# -----------------------------
rf_for_fs = RandomForestClassifier(n_estimators=200, random_state=42, n_jobs=-1)
rf_for_fs.fit(X, y)
importances = pd.Series(rf_for_fs.feature_importances_, index=X.columns).sort_values(ascending=False)

threshold = np.median(importances.values)
sel = SelectFromModel(rf_for_fs, threshold=threshold, prefit=True)
X_selected = sel.transform(X)
selected_features = X.columns[sel.get_support()].tolist()
print(f"Selected {len(selected_features)} features: {selected_features}")

# -----------------------------
# 6) Balance dataset (SMOTE)
# -----------------------------
if use_smote:
    smote = SMOTE(random_state=42, n_jobs=-1)
    X_bal, y_bal = smote.fit_resample(X_selected, y)
    print("After SMOTE:", X_bal.shape, y_bal.shape)
else:
    X_bal, y_bal = X_selected, y

# -----------------------------
# 7) Scale features
# -----------------------------
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_bal)

# -----------------------------
# 8) Train-test split
# -----------------------------
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y_bal, test_size=0.20, random_state=42, stratify=y_bal
)
print("Train:", X_train.shape, "Test:", X_test.shape)

# -----------------------------
# 9) Base learners
# -----------------------------
estimators = []
estimators.append(("rf", RandomForestClassifier(n_estimators=400, max_depth=12, random_state=42, n_jobs=-1)))
estimators.append(("gb", GradientBoostingClassifier(n_estimators=300, learning_rate=0.05, max_depth=5, random_state=42)))

if use_xgb:
    estimators.append(("xgb", XGBClassifier(
        n_estimators=300, learning_rate=0.05, max_depth=6,
        use_label_encoder=False, eval_metric="logloss",
        random_state=42, n_jobs=1)))
if use_cat:
    estimators.append(("cat", CatBoostClassifier(
        iterations=300, learning_rate=0.05, depth=6,
        verbose=0, random_state=42)))

# -----------------------------
# 10) Stacking model
# -----------------------------
stack = StackingClassifier(
    estimators=estimators,
    final_estimator=LogisticRegression(max_iter=1000),
    n_jobs=-1
)

# -----------------------------
# 11) Train & Evaluate
# -----------------------------
print("\nTraining final model...")
stack.fit(X_train, y_train)

y_pred = stack.predict(X_test)
acc = accuracy_score(y_test, y_pred)
print("\nAccuracy:", round(acc, 4))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
cm = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:\n", cm)

# Plot confusion matrix
plt.figure(figsize=(6,5))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues",
            xticklabels=["No Flood", "Flood"],
            yticklabels=["No Flood", "Flood"])
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix")
plt.savefig("confusion_matrix.png")  # Save as PNG
plt.show()

# ROC-AUC
try:
    y_proba = stack.predict_proba(X_test)[:, 1]
    roc = roc_auc_score(y_test, y_proba)
    print("ROC-AUC:", round(roc, 4))
except:
    pass

# -----------------------------
# 12) Save model artifacts
# -----------------------------
joblib.dump({
    "stack_model": stack,
    "scaler": scaler,
    "encoders": encoders,
    "selected_features": selected_features
}, "flood_stack_artifacts.pkl")

# print("\nModel + scaler + encoders saved as 'flood_stack_artifacts.pkl'")
# print("Confusion matrix plot saved as 'confusion_matrix.png'")
```
